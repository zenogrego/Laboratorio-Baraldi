Sono stati presi in questo caso i valori di pressione ogni  $ 2 $ minuti di tempo. Questo seti di dati è stato raccolto nel file $ \textbf{ data_25_05_22.csv} $.

Importiamo i dati di pressione e tempo:

```python
#importiamo i dati

data = pd.read_csv('/Users/zenogregorin/Desktop/LAB_BARALDI/25_05_22/data_25_05_22.csv')
```

Visualizziamo i dati: 

```python
#visualizziamo i dati

data
```

```python
#definiamo x e y le colonne che ci interessano di tempo e pressione

x = data.Time
y = data.Pressure
```

Visualizziamo in un grafico i dati 

```python
plt.plot(x, y, 'o--', color = 'red', label = 'Data measurement')
plt.xlabel('Time [min]')
plt.ylabel('Pressure [mbar]')
plt.grid()
plt.legend()
plt.show()
```

Andiamo a riproporre lo stesso grafico affiancato al grafico in cui prendiamo l'asse delle y in scala semilogaritmica:

```python
fig, axes = plt.subplots(1,2, figsize=(14, 4))

#grafico sulla sinistra
ax = axes[0] 
ax.plot(x, y, 'o--', color = 'red', label = 'Data measurement')
ax.set_title("Grafico P-t")
ax.set_xlabel("Time [min]")
ax.set_ylabel("Pressure [mbar]")
ax.grid()

#grafico sulla destra
ax = axes[1]
ax.set_yscale('log')
ax.plot(x, y, 'o--', color = 'black', label = 'Data measurement')
ax.set_title("Grafico P-t scala semilog")
ax.set_xlabel("Time [min]")
ax.set_ylabel("Pressure [mbar]")
ax.grid()


plt.show()
```
